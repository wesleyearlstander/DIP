```python
import numpy as np
import matplotlib.pyplot as plt
from skimage import io
```

Read in images

```python
img1 = io.imread("images/chap1&2/girl.tif")
img2 = io.imread("images/chap1&2/boy.tif")
img3_rgb = io.imread("images/chap1&2/lenna-RGB.tif")
img3_gray = io.imread("images/chap1&2/lenna-RGB.tif", as_gray=True)
img4 = io.imread('images/chap1&2/chronometer.tif')
```

Display an image

```python
io.imshow(img4, cmap=plt.cm.gray)
plt.show()
```

Rescale an image

```python
from skimage.transform import rescale
```

```python
img4_rescaled = rescale(img4, 0.25, multichannel=False)
io.imshow(img4_rescaled, cmap='gray')
plt.show()
```

```python
img4_rescaled.dtype
```

Save an image

```python
from skimage import img_as_ubyte
img4_rescaled = img_as_ubyte(img4_rescaled)
```

```python
img4_rescaled.dtype
```

```python
io.imsave('images/chap1&2/chronometer_rescaled.png', img4_rescaled)
```

Display images in rows and columns

```python
fig,axes = plt.subplots(nrows=2, ncols=2, figsize=(8,8), sharex=True, sharey=True)
ax=axes.ravel()
ax[0].imshow(img1, cmap=plt.cm.gray)
ax[0].set_title('Girl')

ax[1].imshow(img2, cmap='gray')
ax[1].set_title('Boy')

ax[2].imshow(img3_rgb)
ax[2].set_title('Lenna-RGB')

ax[3].imshow(img3_gray, cmap=plt.cm.gray)
ax[3].set_title('Lenna-gray')

for a in ax:
    a.axis('off')

fig.tight_layout()
plt.show()
```
