```python
import matplotlib.pyplot as plt
import numpy as np
import math
from skimage import util
from skimage import exposure
import skimage.morphology as m
from skimage.color import rgb2gray
from scipy import ndimage as nd
```

# Question 1

```python
img1 = rgb2gray(plt.imread("images/text-image.tif"))
plt.gray()
plt.imshow(img1)
```

```python
SE = np.ones((51,1))
img1D = m.binary_opening(img1, SE)
plt.imshow(img1D)
```

```python
img1O = np.logical_and(img1D,img1)
plt.imshow(img1O)
```

```python
def extract (img, SE) :
    img1F = m.binary_erosion(img1, SE)
    SE2 = m.square(3)
    imgTemp = np.logical_and(m.dilation(img1F, SE2), img)
    img1D = np.zeros(1)
    while (not (img1D==imgTemp).all()) :
        img1D = np.logical_and(m.dilation(imgTemp, SE2), img)
        imgTemp = np.logical_and(m.dilation(img1D, SE2), img)
    return imgTemp
```

```python
SE = np.ones((51,1))
img1F = extract(img1, SE)
plt.imshow(img1F)
```

# Question 2

```python
img2 = rgb2gray(plt.imread("images/text-image.tif"))
plt.imshow(img2)
```

```python
from skimage import measure
from skimage import util
from skimage.color import label2rgb

SE = np.ones((img2.shape[0]-8, img2.shape[1]-8))
SE = np.pad(SE, pad_width = 4)
SE = util.invert(SE)

img21 = np.logical_and(img2,SE)
plt.imshow(img21)
```

```python
def connectedExtraction (img, imgE):
    SE = m.square(3)
    imgO = np.logical_and(m.dilation(imgE, SE), img)
    imgTemp = np.ones(1)
    while(not (imgO==imgTemp).all()):
        imgTemp = np.logical_and(m.dilation(imgO, SE), img)
        imgO = np.logical_and(m.dilation(imgTemp, SE), img)
    return img-imgO
```

```python
img22 = connectedExtraction(img2, img21)
plt.imshow(img22)
```

# Question 3

```python
img3 = rgb2gray(plt.imread("images/circuitboard_clean.tif"))
plt.imshow(img3)
```

```python
img31 = m.erosion(img3, m.square(3))
img32 = m.dilation(img3, m.square(3))
img33 = m.opening(img3, m.square(3))
img34 = m.closing(img3, m.square(3))
fig, axs = plt.subplots(1,4,figsize=(20,5))
[axi.set_axis_off() for axi in axs.ravel()]
axs[0] = fig.add_subplot(1,4,1)
axs[0].set_title("Erosion")
axs[0].imshow(img31)
axs[1] = fig.add_subplot(1,4,2)
axs[1].set_title("Dilation")
axs[1].imshow(img32)
axs[2] = fig.add_subplot(1,4,3)
axs[2].set_title("Opening")
axs[2].imshow(img33)
axs[3] = fig.add_subplot(1,4,4)
axs[3].set_title("Closing")
axs[3].imshow(img34)
```
