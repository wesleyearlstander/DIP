```python
import matplotlib.pyplot as plt
import numpy as np
import math
from skimage import util
from skimage import exposure
import skimage.morphology as m
from skimage.color import rgb2gray
from scipy import ndimage as nd
```

```python
plt.gray()
img = plt.imread("images/UTK.tif")
img = rgb2gray(img)
plt.imshow(img)
img = 1.0*img
```

```python
img1 = m.erosion(img, m.square(3))
img1 = m.erosion(img1, m.square(3))
img1 = m.erosion(img1, m.square(3))
plt.imshow(img1)
```

```python
img12 = m.dilation(img, m.square(3))
img12 = m.dilation(img12, m.square(3))
img12 = m.dilation(img12, m.square(3))
plt.imshow(img12)
```

```python
img13 = m.erosion(img, m.square(7))
plt.imshow(img13)
```

```python
diff = np.subtract(img13, img1)
plt.imshow(diff)
#each time an erosion of 2 occurs around the central pixel hence after 3 times then 6 occurs + central pixel = 7
```

```python
img14 = util.invert(img)
plt.imshow(img14)
```

```python
img15 = m.erosion(img14, m.square(9))
plt.imshow(img15)
#the white pixels surrounding the focus are eroded away which looks like a dilation of black pixels
```
