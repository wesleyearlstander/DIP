```python
import matplotlib.pyplot as plt
img = plt.imread ('images/girl.tif') #only png supported to change
print(img.dtype) #image type
print(img.shape) #image shape
plt.gray ()
plt.imshow (img)
plt.axis ("on")
```

```python
def PrintMiddlePixel (image):
    x = (int)(img.shape[0]/2 - 1)
    y = (int)(img.shape[1]/2 - 1)
    print(image[x][y])
```

```python
PrintMiddlePixel(img)
```

```python
def scanLine(f, I, loc):
    out = [0]
    if (loc == "row"):
        out = [0] * f.shape[1]
        for i in range(f.shape[1]):
            out[i] = f[I][i]
    else:
        out = [0] * f.shape[0]
        for i in range(f.shape[0]):
            out[i] = f[i][I]
    return out
```

```python
hLine = scanLine(img, (int)(img.shape[0]/2), "row")
fig = plt.figure()
ax = plt.axes()
ax.plot(hLine)
```

```python
img2 = plt.imread ('images/spillway-dark.tif')
print(img2.dtype)
print(img2.shape) 
plt.gray ()
plt.imshow (img2)
plt.axis ("on")
```

```python
import numpy as np
```

```python
def contrast(image):
    return np.max(image) - np.min(image)
```

```python
print("[" + str(np.min(img2)) + "," + str(np.max(img2)) + "]")
```

```python
def ScaleIntensity (image, K):
    m = contrast(image)
    out = np.array(image).astype(float)
    for j in range(image.shape[1]):
        for i in range(image.shape[0]):
            out[i][j] = K*(out[i][j]/m)
    return out
```

```python
img3 = ScaleIntensity(img2, 255)
img4 = ScaleIntensity(img2, 1)
```

```python
plt.imshow(img3)
```

```python
plt.imshow(img4)
```
