# Question 2

```python
img21 = util.invert(img1)
img22 = m.erosion(img21, m.disk(2))
img23 = img21-img22
plt.imshow(img23)
```

The images are similar but there are differences, the second image is closer to the actual boundary as the foreground of the image was actually the background pixels in the original image so the compliment provides a better boundary extraction as it is the internal boundary and the first image is the external boundary

# Question 3

```python
img3 = plt.imread("images/small_squares.png")
plt.imshow(img3)
```

```python
SE1 = np.ones((4,4))
SE2 = np.zeros((4,4))
SE2 = np.pad(SE2, [(1, 1), (1, 1)], mode='constant', constant_values=(1,1))

img31 = nd.morphology.binary_hit_or_miss(img3, SE1, SE2)
plt.imshow(img31)
```

```python
img32 = m.dilation(img31, SE1)
plt.imshow(img32)
```

The process involves detecting the object using a hit-or-miss operation then dilating the pixel using the original shape

# Question 6

```python
img6 = rgb2gray(plt.imread("images/UTK.tif"))
plt.imshow(img6)
```

```python
from skimage import measure
from skimage.color import label2rgb
img61 = measure.label(img6)
image_label_overlay = label2rgb(img61, image=img6, bg_label=0)
plt.imshow(image_label_overlay)
```

```python
def highlight(f,intensityRange, preserveBackground = False):
    out = np.copy(f)
    for i in range(len(f)):
        for j in range(len(f[0])):
            if f[i][j]>=intensityRange[0] and f[i][j]<=intensityRange[1]:
                out[i][j] = 255
            elif preserveBackground == False:
                out[i][j] = 0
    return out
```

```python
img6U = highlight(img61, [0.5,1.5])
plt.imshow(img6U)
```

```python
img6T = highlight(img61, [1.5,2.5])
plt.imshow(img6T)
```

```python
img6K = highlight(img61, [2.5,3.5])
plt.imshow(img6K)
```
