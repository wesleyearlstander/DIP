```python
import matplotlib.pyplot as plt
import numpy as np
import math
import skimage.filters as f
from skimage import exposure
import skimage.io as io
from PIL import Image
```

```python
def createBoxFilter(size):
    out = np.zeros((size,size))
    for i in range(size):
        for j in range(size):
            out[i][j] = 1/(size*size)
    return out
```

```python
def gausskernel(m,sig,K=1):
    out = np.zeros((m,m))
    total = 0
    for i in range(m):
        for j in range(m):
            r = ((i-(m-1)/2)**2) + ((j-(m-1)/2)**2)
            out[i][j] = K * math.exp(-1 * (r/(2*(sig**2))))
            total += out[i][j]
    return out/total;        
```

```python
def twodConv(f,w):
    m = len(w)
    n = len(w[0])
    if m == n:
        y = len(f)
        x = len(f[0])
        g = np.pad(f, int((m-1)/2), mode='reflect')
        new = np.zeros((y,x))
        for i in range(y):
            for j in range(x):
                    new[i][j] = np.sum(g[i:i+m, j:j+m] * w)
        return new
```
